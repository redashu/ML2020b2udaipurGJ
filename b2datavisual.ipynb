```
# in case of using lap/PC 
#  pip  install matplotlib  seaborn   # in PC / laptop 
#!pip  install matplotlib  seaborn  # in colab 
x=[2,5,4,9]
y=[3,8,3,6]
```

```
import  matplotlib.pyplot   as plt
```

```
plt.xlabel('distance') # for labeling x axis
plt.ylabel('time')  #   for y 
plt.scatter(x,y,label="minsss")  #  to show dots only  
#plt.scatter(y,x)
plt.plot(x,y,label="railway track")   #  to plot line 
plt.bar(x,y,label="finace growth") # to show bar plot
plt.bar(y,x,color='yellow')
plt.grid(c='green')  #  for  grid  and color as well
plt.legend()  #  to show labels of  plot area
plt.plot()  # to plot the graph 
```

```
#  cricket  score  graph
players=["dhoni","rohit","virat"]
runs=[120,160,130]
```

```
plt.xlabel("players")
plt.ylabel("runs")
#plt.bar(players,runs,label="2020 starting match")
plt.pie(runs,labels=players,autopct='%1.1f%%',explode=(0,0.3,0),shadow=True)
plt.legend()
plt.show()
```
